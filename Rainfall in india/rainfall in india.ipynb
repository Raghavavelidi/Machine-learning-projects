```python

c['JAN']=c['JAN'].fillna(c['JAN'].mean())
c['FEB']=c['FEB'].fillna(c['FEB'].mean())
c['MAR']=c['MAR'].fillna(c['MAR'].mean())
c['APR']=c['APR'].fillna(c['APR'].mean())
c['MAY']=c['MAY'].fillna(c['MAY'].mean())
c['JUN']=c['JUN'].fillna(c['JUN'].mean())
c['JUL']=c['JUL'].fillna(c['JUL'].mean())
c['AUG']=c['AUG'].fillna(c['AUG'].mean())
c['SEP']=c['SEP'].fillna(c['SEP'].mean())
c['OCT']=c['OCT'].fillna(c['OCT'].mean())
c['NOV']=c['NOV'].fillna(c['NOV'].mean())
c['DEC']=c['DEC'].fillna(c['DEC'].mean())
```

```python

c.isna().sum()
```

```python
c['ANNUAL RAIN']=c.apply(lambda row: row.JAN + row.FEB,axis=1)
c
```

```python
d=c.drop(['SUBDIVISION'],axis=1)
```

```python
cor=d.corr()
cor
```

```python

import seaborn as sb
sb.heatmap(cor,vmax=1,vmin=-1,annot=True,linewidth=-5,cmap="bwr")
```
