```python
data6=data.loc[(data.ref_model =="i10")]
data6.head(10)
```

```python
data7=data.loc[(data.ref_model =="Swift")]
data7
```

```python
data8=data.loc[(data.ref_model =="Wagon")]
data8
```

```python
a=pd.concat([data6,data7,data8])
a
```

```python
data1=a.drop(['city','model_name','fuel_type'],axis=1)
data1
```

```python
data1["pre_owner"] = data1['pre_owner'].map({"1st Owner":1,"2nd Owner":2,"3rd Owner":3,"4th Owner":4,"6th Owner":6})
```

```python
data1
```

```python
data2=pd.get_dummies(data1,dtype=int)
data2
```

```python
y=data2['price (₹)']
x=data2.drop('price (₹)',axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
ypred=reg.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(ypred,y_test)
```

```python
mse
```
